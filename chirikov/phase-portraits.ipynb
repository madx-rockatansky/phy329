```python
# standard Python libraries
import numpy as np
import matplotlib.pyplot as plt
# project modules
from map.standardMap import StandardMap
from plots.mapPlot import plot_phase_generic
```

```python
# small K: mostly tori
K = 0.2
n_iters = 5000
n_sim = 25

m = StandardMap(K=K, nIters=n_iters, seed=1)
m.simulate(ic=n_sim)
run = m.runs[-1]["run"]

plot_phase_generic(
    run=run,
    mode="phase",
    point_size=0.15,
    title=rf"Phase Space for $K = {K}$ ({n_sim} ICs)"
)
```
